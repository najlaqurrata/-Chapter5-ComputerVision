### Import libraries

```python
import numpy as np, cv2
from skimage.io import imsave
from skimage.measure import block_reduce
from torch.nn import MaxPool2d
from torch import from_numpy
from matplotlib import pyplot as plt
from matplotlib.image import imread
```

### User-defined Functions (UDFs)

```python
def plot(
    imgs : list,
    titles : list = ["ORIGINAL", "EDITED"],
    figsize : tuple = (8, 5),
    cmaps : list = [None, None],
    *args, **kwargs
) -> None:

    if len(imgs) > 3:
        raise ValueError("Max. total image = 3.")

    if len(imgs) != len(titles):
        raise ValueError("'imgs' should be in the same length with 'titles'.")

    my_figure = plt.figure(figsize=figsize)
    # NOTE: figsize=(width/horizontally, height/vertically)

    fig_ids = [121, 122] if len(imgs) == 2 else [131, 132, 133]

    for idx, img, title, c in zip(fig_ids, imgs, titles, cmaps):
        m = my_figure.add_subplot(idx)
        plt.imshow(img, cmap=c)
        plt.axis("off")
        plt.title(title)

    plt.tight_layout()
```

```python
def apply_clahe_rgb(img, clipLimit=2, tileGridSize=(10, 10)):
    # LAB color model where L = Lightness, a = green-red, b = blue-yellow
    light, a, b = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2LAB)) # convert to LAB color model
    light_clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize).apply(light) # apply CLAHE for lightness (L) component only
    return cv2.cvtColor(cv2.merge((light_clahe, a, b)), cv2.COLOR_LAB2BGR) # return with reconvert to BGR color model
```

### Max Pooling (using Scikit-image)

```python
BLOCK_SIZE = (4, 4, 1)
```

```python
img_edited = block_reduce(img, block_size=BLOCK_SIZE, func=np.max)
plot(imgs=[img, img_edited])
```

### Max Pooling (using PyTorch)

```python
# convert numpy-based image to torch.Tensor type
img_torch = from_numpy(img).permute(2, 0, 1).float()

# prepare Max Pooling operation from PyTorch
max_pooling = MaxPool2d(kernel_size=9, stride=2)

# perform Max Pooling operation
img_torch_edited = max_pooling(img_torch).permute(1, 2, 0).int().numpy()
plot(imgs=[img, img_torch_edited])
```

Both PyTorch and scikit-image are well-known libraries in the fields of computer vision and image processing, although they have distinct purposes and functions. Where PyTorch serves as a deep learning framework, and scikit-image serves as an image processing library.

In terms of the mathematical operation performed (selecting the maximum value from a set of values), PyTorch and scikit-image both offer similar functionality for max pooling operations. However, there are distinctions in how they are implemented and the context in which they are utilized.

Pytorch Maxpooling :
1.  Integration with Neural Networks.
2.  Supports automatic differentiation.
3.  GPU acceleration for enhanced performance.

Scikit-Image Maxpooling :
1. Primarily for general image processing.
2. Numpy-based operations.
3. Emphasizes Ease of Use.

In summary, while scikit-image's max pooling operation is well-suited for general image processing tasks with a focus on ease of use, PyTorch's implementation offers additional advantages tailored to the specific requirements of deep learning applications. These advantages include seamless integration with neural networks, automatic differentiation for efficient training, and the ability to leverage GPU acceleration for enhanced computational performance. The choice between the two depends on the specific requirements of the task at hand, whether it's deep learning-centric or more broadly focused on image processing.

### Min Pooling and Average Pooling

```python
BLOCK_SIZE1 = (2, 2, 1)
```

```python
img_edited_min = block_reduce(img, block_size=BLOCK_SIZE1, func=np.min) # TODO: Recall the block_reduce(), but with func=np.min
plot(imgs=[img, img_edited_min])
```

```python
img_edited_avg = block_reduce(img, block_size=BLOCK_SIZE1, func=np.mean).astype(int) # TODO: Recall the block_reduce().astype(int), but with func=np.mean
plot(imgs=[img, img_edited_avg])
```

```python
# QUESTIONS
# TODO: What is the difference between Min Pooling and Average Pooling operations?
# If you cannot see the difference, try with another image or change the BLOCK_SIZE.
```

Min pooling involves selecting the minimum value from a window of elements within an image or feature map. This operation is particularly useful for preserving essential features and edges in the data. It can be robust in scenarios where small changes need to be detected, but it can also be sensitive to outliers or extreme values in the data.

On the other hand, average pooling calculates the mean value within a similar window of elements. This operation provides a smoothed representation of the data and is less sensitive to outliers compared to min pooling. Average pooling tends to reduce noise and variations in the data, resulting in a more stable representation.

In summary, the key difference lies in their effects on the data. Min pooling selects the minimum value from a window of elements, preserving essential features and edges. It can be robust to small changes but sensitive to outliers. Average pooling calculates the mean, providing a smoothed representation and being less sensitive to outliers. Min pooling preserves edges, while average pooling smoothens data. The choice depends on the specific task and data characteristics. Max pooling captures significant features, average pooling smoothens, and min pooling is suitable for edge preservation and noise reduction.

### CLAHE
Now imagine you have a better approach to compete Apple's Photonic Engine and Samsung's Adaptive Tetra-squared Pixel technologies using **Contrast Limited Adaptive Histogram Equalization (CLAHE)**. How's the result?

```python
img_clahe = apply_clahe_rgb(img) # TODO: Call the function apply_clahe_rgb() and pass the variable image of "photo1.jpeg"
plot(imgs=[img, img_edited, img_clahe], titles=["ORIGINAL", "MAX-POOLING", "CLAHE"], figsize=(12, 5), cmaps=[None]*3)
```

```python
# QUESTIONS
# TODO: What are the advantages of using CLAHE (instead of Max Pooling) to brighten the dark-toned image?
```

The advantages of utilizing CLAHE (Contrast Limited Adaptive Histogram Equalization) over Max Pooling for brightening dark-toned images are noteworthy. CLAHE stands out as an image processing technique renowned for its ability to enhance local contrast, particularly in regions with low luminosity, ensuring the preservation of crucial details that might be lost in the shadows. Its adaptive nature allows CLAHE to adjust to varying lighting conditions, thereby mitigating information loss and contributing to an overall improvement in image quality.

On the other hand, max pooling serves a different purpose, primarily utilized in the context of neural networks for down-sampling. Its main function is to reduce the spatial dimensions of feature maps, aiding in feature extraction and computational efficiency during deep learning tasks.

In summary, CLAHE and max pooling are image processing techniques with unique advantages. CLAHE enhances local contrast, especially in dark areas, preserving crucial details. It adapts to varying lighting conditions, reducing information loss, and improving overall image quality. Max pooling, on the other hand, is primarily used for down-sampling in neural networks. The choice between CLAHE and max pooling depends on specific task requirements and desired outcomes.
